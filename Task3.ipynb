```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv("Titanic-Dataset.csv")
```

**3. Data Cleaning and Preprocessing**

*   Data Cleaning and Handling Missing Values
*   Data Preprocessing for Titanic Dataset
*   Data Cleaning Phase of Exploratory Data Analysis (EDA)
*   Preparing the Dataset for Analysis
*   Initial Data Cleaning and Preparation


```python
df['Age'] = df['Age'].fillna(df['Age'].mean())

df = df.drop(columns=['Cabin'], errors='ignore')

df['Embarked'] = df['Embarked'].fillna(df['Embarked'].mode()[0])

print("Missing values after cleaning:\n")
print(df.isnull().sum())
```

**4. Feature Engineering**

*   Feature Engineering and Data Transformation
*   Creating New Features for Analysis
*   Feature Creation (Age Groups and Family Size)
*   Enhancing Dataset with Derived Features








```python
df['AgeGroup'] = pd.cut(
    df['Age'],
    bins=[0, 12, 18, 30, 50, 80],
    labels=['Child', 'Teen', 'YoungAdult', 'Adult', 'Senior']
)


df['FamilySize'] = df['SibSp'] + df['Parch']
```

**5. Exploratory Data Analysis (EDA) and Visualization**

Age Distribution

```python
plt.figure()
sns.histplot(df['Age'], kde=True)
plt.title("Age Distribution")
plt.show()
```

 Survival Rate by Age Group

```python
plt.figure()
sns.barplot(x='AgeGroup', y='Survived', data=df)
plt.title("Survival Rate by Age Group")
plt.show()
```

Survival by Embarkation Port

```python
plt.figure()
sns.barplot(x='Embarked', y='Survived', data=df)
plt.title("Survival Rate by Embarkation Port")
plt.show()
```

 Survival by Family Size

```python
plt.figure()
sns.barplot(x='FamilySize', y='Survived', data=df)
plt.title("Survival Rate by Family Size")
plt.show()
```

Correlation Heatmap

```python
plt.figure()
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```
